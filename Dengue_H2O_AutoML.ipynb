```
from h2o.automl import H2OAutoML
```

```
df = h2o.import_file('/content/Dengue-Dataset.csv')
```

```

churn_train,churn_test,churn_valid = df.split_frame(ratios=[.7, .15])
churn_train,churn_test,churn_valid
```

```
churn_train
```

```
y = "Gender"
x = list(df.columns)
x.remove(y)  # Remove the target column from the predictor list
```

```
aml = H2OAutoML(max_models = 10, seed = 10, exclude_algos = ["StackedEnsemble", "DeepLearning"], verbosity="info", nfolds=0)
aml
```

```

churn_train, churn_test, churn_valid = df.split_frame(ratios=[.7, .15], seed=123)
aml.train(x, y, training_frame=churn_train, validation_frame=churn_valid)
```

```
lb = aml.leaderboard
lb
```

```
lb.head(10)
```

```
churn_pred=aml.leader.predict(churn_test)
```

```
churn_pred.head(10)
```

```
aml.leader.model_performance(churn_test)
```
