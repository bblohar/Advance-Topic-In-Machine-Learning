```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Reshape
```

```python
np.random.seed(42)
tf.random.set_seed(42)
codings_size=100
```

```python
generator=Sequential()
generator.add(Dense(100,activation="relu",input_shape=[codings_size]))
generator.add(Dense(150,activation="relu"))
generator.add(Dense(28*28,activation="sigmoid"))
generator.add(Reshape([28,28]))
```

```python

discrimator=Sequential()
discrimator.add(Flatten(input_shape=[28,28]))
discrimator.add(Dense(150,activation="relu"))
discrimator.add(Dense(100,activation="relu"))
discrimator.add(Dense(1,activation="sigmoid"))
discrimator.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
GAN=Sequential([generator,discrimator])
```

```python
discrimator.trainable=False
```

```python
GAN.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
import tensorflow as tf
```

```python
batch_size=32
```

```python
my_data=only_zeros
```

```python
my_data.shape
```

```python
dataset=tf.data.Dataset.from_tensor_slices(my_data).shuffle(buffer_size=1000)
```

```python
num_elements=sum(1 for _ in my_data)
print("number os data",num_elements)
```

```python
num_elements=dataset.cardinality().numpy()
```

```python
dataset=dataset.batch(batch_size,drop_remainder=True).prefetch(1)
```

```python
epochs=10
```

```python
for epoch in range(epochs):

    print(f"Epoch {epoch+1}")

    i=0

    for x_batch in dataset:

      i=i+1

      if i%10 == 0:

        print(f"\t current on batch number",{epoch+1})

        noise=tf.random.normal(shape=[batch_size,codings_size])

        generated_images=generator(noise)

        # Convert x_batch to float32 before concatenation

        x_batch = tf.cast(x_batch, tf.float32)

        x_fake_and_real=tf.concat([generated_images,x_batch],axis=0)

        y1=tf.constant([[0.]]* batch_size+[[1.]]*batch_size)

        discrimator.trainable=True

        discrimator.train_on_batch(x_fake_and_real,y1)

        noise=tf.random.normal(shape=[batch_size,codings_size])

        y2=tf.constant([[1.]]*batch_size)

        discrimator.trainable=False

        GAN.train_on_batch(noise,y2)

      print("training complete")
```
